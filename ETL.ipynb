```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.optimizers import RMSprop
```

```python
raw_data = pd.read_csv("data/exoplanet_data.csv")
```

```python
display(raw_data)
display(raw_data.info())
display(raw_data.isna().any())
```

# Prepare data for training

```python
raw_data['koi_disposition'].value_counts()
```

```python
# Drop rows where there is not a definitive classification
learning_data = raw_data[~raw_data.koi_disposition.isin(['CANDIDATE'])]
display(learning_data['koi_disposition'].value_counts())
```

```python
learning_data.to_csv("data/learning_data.csv")
```

# Proof of concept using candidate verification tests for training

```python
# Select data features to train on, and the column of labels row each row of data
exoplanet_data = learning_data[['koi_fpflag_nt','koi_fpflag_ss','koi_fpflag_co','koi_fpflag_ec']]
exoplanet_labels = learning_data['koi_disposition'].replace({
    'FALSE POSITIVE':0,
    'CONFIRMED':1
}).to_frame() 

print(exoplanet_data.shape, exoplanet_labels.shape)
```

```python
display(exoplanet_data)
display(exoplanet_labels)
```

```python
# Split the data into training and validation sets
training_data, validation_data, \
training_labels, validation_labels = train_test_split(
    exoplanet_data,
    exoplanet_labels,
    random_state=1
)

print(training_data.shape)
print(training_labels.shape)
print(validation_data.shape)
print(validation_labels.shape)
```

```python
# build and compile the neural network
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(16, activation = 'relu',input_dim=4),
    tf.keras.layers.Dense(, activation = 'softmax')
])

model.compile(
    optimizer=RMSprop(learning_rate=.01),
    loss='categorical_crossentropy',
    metrics=['accuracy'])
```

```python
history = model.fit(
    training_data,
    training_labels,
    validation_data = (validation_data, validation_labels),
    epochs=100,
    batch_size=50)
```
